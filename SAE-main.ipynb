```python
import json
import requests
from bs4 import BeautifulSoup
def download_page(url):
    response = requests.get(url)
    response.raise_for_status()
    return response.text
def main(url):
    content = download_page(url)
    soup = BeautifulSoup(content, 'html.parser')
    result = {}
    for row in soup.table.find_all('tr'):
        row_header = row.th.get_text()
        row_cell = row.td.get_text()
        result[row_header] = row_cell
    with open('book_table.json', 'w') as storage_file:
        storage_file.write(json.dumps(result))
if __name__ == "__main__":
    main("https://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
import pandas as pd


#paths
path_name = "//*[@id='app']/section/div/div[1]/div[2]/div[1]/div"
path_director = "//*[@id='app']/section/div/div[1]/div[4]/div[1]/div/div[2]/div[1]/div[1]/dl/dd[1]"
path_dependencia = "//*[@id='app']/section/div/div[1]/div[4]/div[1]/div/div[2]/div[1]/div[1]/dl/dd[2]"
path_niveles=  "//*[@id='app']/section/div/div[1]/div[4]/div[1]/div/div[2]/div[1]/div[1]/dl/dd[3]"
path_orientacion = "//*[@id='app']/section/div/div[1]/div[4]/div[1]/div/div[2]/div[1]/div[1]/dl/dd[3]"
path_direccion="//*[@id='app']/section/div/div[1]/div[4]/div[2]/div/div[1]/div[2]/div[1]/div[1]/div[2]/a/u"
path_region = "//*[@id='app']/section/div/div[1]/div[4]/div[2]/div/div[1]/div[2]/div[1]/div[2]/div[2]/span"
path_comuna = "//*[@id='app']/section/div/div[1]/div[4]/div[2]/div/div[1]/div[2]/div[1]/div[3]/div[2]/span"
path_telefono = "//*[@id='app']/section/div/div[1]/div[4]/div[2]/div/div[1]/div[2]/div[2]/div[1]/dl/a/dd/div"
path_pagina = "//*[@id='app']/section/div/div[1]/div[4]/div[2]/div/div[1]/div[2]/div[2]/div[2]/dl/dd/div"
path_error = "//*[@id='modalError-1___BV_modal_body_']"


def getData(i,res):

    name=""
    director=""
    dependencia=""
    niveles=""
    orientacion=""
    direccion=""
    region=""
    comuna=""
    telefono=""
    pagina=""

    
#     er = WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_error)))
    
    
    try:
        name =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_name)))
        name=name.text
    except TimeoutException:
        print("TimeoutException on name: "+str(i))
        return
    except:
        print("Error on name: "+str(i))
    
    try:
        director =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_director)))
        director=director.text
    except TimeoutException:
        print("TimeoutException on director: "+str(i))
    except:
        print("Error on name: "+str(i))

    try:
        dependencia =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_dependencia)))
        dependencia=dependencia.text
    except TimeoutException:
        print("TimeoutException on dependencia: "+str(i))
    except:
        print("Error on name: "+str(i))
    
    try:    
        niveles =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_niveles)))
        niveles = niveles.text
    except TimeoutException:
        print("TimeoutException on niveles: "+str(i))
    except:
        print("Error on name: "+str(i))

    try:
        orientacion =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_orientacion)))
        orientacion=orientacion.text
    except TimeoutException:
        print("TimeoutException on orientacion: "+str(i))
    except:
        print("Error on name: "+str(i))

    try:
        direccion =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_direccion)))
        direccion = direccion.text
    except TimeoutException:
        print("TimeoutException on direccion: "+str(i))
    except:
        print("Error on name: "+str(i))
    
    try:
        region =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_region)))
        region = region.text
    except TimeoutException:
        print("TimeoutException on region: "+str(i))
    except:
        print("Error on name: "+str(i))

    try:
        comuna =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_comuna)))
        comuna = comuna.text
    except TimeoutException:
        print("TimeoutException on comuna: "+str(i))
    except:
        print("Error on name: "+str(i))

    
    try:
        telefono =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_telefono)))
        telefono = telefono.text
    except TimeoutException:
        print("TimeoutException on telefono: "+str(i))
    except:
        print("Error on name: "+str(i))    

    try:
        pagina =  WebDriverWait(driver, TIME_OUT).until(EC.visibility_of_element_located((By.XPATH, path_pagina)))
        pagina = pagina.text
    except TimeoutException:
        print("TimeoutException on pagina: "+str(i))
    except:
        print("Error on name: "+str(i))
    
    fila = [i,name,director,niveles,dependencia,orientacion,direccion,region,comuna,telefono,pagina]
    res.append(fila)


    
def printToFile(res,header,filename='Query.csv'):
    df = pd.DataFrame(res)
    df.to_csv(filename,index=False,header=header,encoding='latin-1')
    print("Finished!")
    
    
    
    
INT_MAX_COLEGIOS = 30000
TIME_OUT = 2
URL_BASE = 'https://admision.mineduc.cl/vitrina-vue/establecimiento/'
ERROR_MESSAGE = "Error on"
header = ['Nro','Nombre','Director','Niveles','Dependencia','Orientacion','Direccion','Region','Comuna','Telefono','Pagina']


res = []



options = Options()
options.headless = True
driver = webdriver.Chrome(options = options)


for i in range(1,INT_MAX_COLEGIOS):
    URL = URL_BASE + str(i)
    driver.get(URL)
    getData(i,res)
    
printToFile(res,header,'Res-Query.csv')
driver.quit()
```
